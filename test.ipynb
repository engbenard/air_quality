```python
def insertion_sort_tt(arr):
  for i in range(len(arr)):
    key = arr[i]
    current_index = i-1
    while key > arr[current_index] and current_index > -1 :
      arr[current_index+1] = arr[current_index]
      arr[current_index] = key
      current_index = current_index-1
  return arr
```

```python
insertion_sort_tt([4,3,1,2,10,-3])
```

```python
numbers = [1,2,3,5]
def get_missing_numbers(numbers):
  set_numbers = set(range(1,6))
  numbers_to_set = set(numbers)
  return set_numbers-numbers_to_set
get_missing_numbers(numbers)
```

```python
def get_duplicate_strings(arr):
  duplicates, seen = set(), set()
  for i in arr:
    if i in seen:
      duplicates.add(i)
    else:
      seen.add(i)
  return duplicates
m = "Muhumuza Joshua is my name my".split(" ")
get_duplicate_strings(m)
```

```python
# computing intersection of two lists
def intersection_two_lists(l1, l2):
  dups = []
  for i in l1:
    if i in l2:
      dups.append(i)
  return dups
```

```python
def is_anagram(st1, st2):
  if set(st1) == set(st2):
    return True
  return False

is_anagram("rasp","sarp")
```
